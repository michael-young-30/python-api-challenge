```python
import requests
import json
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import time
from scipy.stats import linregress

lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
from citipy import citipy 

from keys import weather_api_key
```

```python
lat_lngs = []
cities=[]

lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)


lat_lngs = zip(lats, lngs)

for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    if city not in cities:
        cities.append(city)
len(cities)
```

```python
url="http://api.openweathermap.org/data/2.5/weather?"
query_url=f"{url}appid={weather_api_key}&q={cities}&units=imperial"
query_url
```

```python
weather_df=pd.DataFrame({
    "City": cities,
    "Latitude": '',
    "Longitude":'',
    "Temperature (F)":'',
    "Humididity (%)" : '',
    "Cloudiness (%)" : '',
    "Wind Speed (mph)" : ''    
})
weather_df
```

```python
query=f"{url}appid={weather_api_key}&q={cities[0]}&units=imperial"
response=requests.get(query).json()
temp=response["main"]["temp"]
temp
```

```python
# query=f"{url}appid={weather_api_key}&q={cities}&units=imperial"

units = "imperial"

# query_url = f"{url}appid={weather_api_key}&units={units}&q="
# print(query_url)

query_url = f"{url}appid={weather_api_key}&units={units}&q="
city=cities
```

```python
number=1
num_set =1

for index,row in weather_df.iterrows():
    city=row['City']
    response = requests.get(query_url+city).json()
    try:
        print(f'Processing Weather Data for Record {number} of Set {num_set} | {response["name"]}')
        weather_df.loc[index,"Temperature (F)"]=(response["main"]["temp"])
        weather_df.loc[index,"Latitude"]=(response["coord"]["lat"])
        weather_df.loc[index,"Longitude"]=(response["coord"]["lon"])
        weather_df.loc[index,"Humididity (%)"]=(response["main"]["humidity"])
        weather_df.loc[index,"Cloudiness (%)"]=(response["clouds"]["all"])
        weather_df.loc[index,"Wind Speed (mph)"]=(response["wind"]["speed"])
        
        number += 1
    except (KeyError, IndexError):
        print(f'not found...skipping')
    if number == 51:
        num_set += 1
        number = 1
        time.sleep(60)
print(f'this works!')
```

```python
weather_df['Temperature (F)'] = weather_df['Temperature (F)'].replace('', np.nan)
city_weather_df = weather_df.dropna()
city_weather_df.tail()
```
